```python
import random
import pandas as pd
from datetime import datetime

#구글 드라이브와 연동
from google.colab import drive
drive.mount('/content/gdrive')
directory_path = '/content/gdrive/My Drive/Colab Notebooks/'

import pandas as pd

data = pd.read_csv(directory_path+'list.csv', encoding = 'UTF-8')

team_list=[]

print("How much team?")
team = int(input())

print("How many person per team?")
max_person = int(input())

grade1 = []
grade2 = []
for i in range (len(data)):
  if data['grade'][i] == 1:
    grade1.append(data['name'][i])
  else:
    grade2.append(data['name'][i])

dt = datetime.now()
random.seed(dt.microsecond)
random.shuffle(grade1)
dt = datetime.now()
random.seed(dt.microsecond)
random.shuffle(grade2)

for i in range (0,team):
  team_list.append([])

for i in range (len(grade1)):
  team_list[i] = [grade1[i]]

for i in range (len(grade2)):
  if i >= team:
    team_list[i-team].append(grade2[i])
  else:
    team_list[i].append(grade2[i])

for i in range (len(team_list)):
  print('team', 2+1, ' : ', team_list[i])
```
