```python
import chromadb
client = chromadb.PersistentClient()
```

```python
results = client.delete_collection(name="posts")
results
```

```python
posts = client.create_collection(
    name="posts"
)
```

```python
posts.delete(ids=["1","2","3","4"])

post1 = 'Apple is delicious'
post2 = 'Banana is sweet'
post3 = 'New York is big'
post4 = 'Paris is romantic'

posts.add(
    documents=[post1, post2, post3, post4],
      ids=["1", "2", "3", "4"]
)
```

```python
result = posts.query(
    query_texts=['yellow'],
    n_results=2
)

result
```

```python
from chromadb.utils import embedding_functions

default_ef = embedding_functions.DefaultEmbeddingFunction()
default_ef(texts=["Banana is sweet"])
```

```python
posts_default_ef = client.create_collection(
    name="posts_default",
    embedding_function=default_ef
)
```
